<h4>Step 3:- Handle Null Values</h4>

```python
matches.isnull().sum()[matches.isnull().sum()>0]
```

```python
matches['city'] = matches['city'].fillna('unknown')
```

<h4>Step 4:- Team wise Wins analysis</h4>

```python
team_wins = matches['winner'].value_counts()
plt.figure()
team_wins.plot(kind='bar')
plt.title('Most Successful IPL Teams')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.show()
```

<h4>Step 5:- Matches per Season</h4>

```python
matches['date'] = pd.to_datetime(matches['date'])
```

```python
matches['season'] = matches['date'].dt.year
```

```python
season_matches = matches['season'].value_counts().sort_index()
plt.figure()
season_matches.plot(kind='line',marker='o')
plt.title('Matches Played per Season')
plt.xlabel('Season')
plt.ylabel('Matches')
plt.show()
```

<h4>Step 6:- Top Run Scorers</h4>

```python
top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
plt.figure()
top_batsmen.plot(kind='bar')
plt.title('Top 10 Run Scores in IPL')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```

<h4>Step 7:- Top Wicket Takers</h4>

```python
wickets = deliveries[deliveries['dismissal_kind'].notnull()]
top_bowlers = wickets['bowler'].value_counts().head(10)
plt.figure()
top_bowlers.plot(kind='bar')
plt.title('Top 10 Wicket Takers in IPL')
plt.xlabel('Bowker')
plt.ylabel('Wickets')
plt.show()
```

<h4>Step 8:- Toss Decision Analysis</h4>

```python
toss_decision = matches['toss_decision'].value_counts()
plt.figure()
toss_decision.plot(kind='pie',autopct='%1.1f%%')
plt.title('Toss Decision Distribution')
plt.ylabel('')
plt.show()
```

<h4>Step 9:- Venue Wise Matches</h4>

```python
top_venues = matches['venue'].value_counts().head(10)
plt.figure()
top_venues.plot(kind='barh')
plt.title('Top 10 IPL Venues by Matches Hosted')
plt.show()
```

<h4>Step 10:- Seaborn Visualizations</h4>

```python
sns.set(style='whitegrid')
plt.figure()
sns.countplot(data=matches, y='winner',
             order=matches['winner'].value_counts().index)
plt.title('Team wise IPL Wins')
plt.show()
```
